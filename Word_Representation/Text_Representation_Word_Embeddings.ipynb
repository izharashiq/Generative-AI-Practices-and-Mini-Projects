## Bag-of-Words

```python
bow = cv.fit_transform(df['text'])
```

# Ngrams

```python
df = pd.DataFrame({
    "text": [
        "The cat sat on the mat",
        "Cats are very curious animals",
        "The quick brown fox jumps over the lazy dog.",
        "Dogs are loyal and friendly pets",
        "I love reading books on a rainy day",
        "Artificial Intelligence is transforming the future of technology",
        "Machine Learning and AI are closely related fields",
        "Python is a popular programming language for data science",
        "Data visualization helps make complex information easy to understand",
        "Deep learning techniques are revolutionizing image and language processing"
    ],
    "output": [1, 1, 0, 0, 1, 1, 1, 1, 0, 0]
})

df
```

```python
# BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
# Ti gram
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(3,3))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
bow = bow.toarray()
```

```python
bow
```
